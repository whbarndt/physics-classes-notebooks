# Taylor series

- Find a general method for writing a power series respresentation
  for a function.
- Assume that $f(x)$ does in fact have a power series
  representation about $x=a$, i.e., 
  
\begin{equation*}
f(x) = \sum \limits_{n=0}^{\infty} c_n(x-a)^n = c_0 + c_1(x-a) + c_2(x-a)^2 + \cdot \cdot \cdot
\end{equation*}

- Assume $f(x)$ has derivatives of every order.
- To find $c_0$, just evaluate everything at $x=a$, 


\begin{equation*}
f(a) = c_0
\end{equation*}

- How do we find the other coefficients?  Take the derivative of
  the function (and its power series) and evaluate at $x=a$.  Here's
  the first derivative...

\begin{equation*}
f^\prime(x) = c_1 + 2 c_2(x-a) + 3 c_3(x-a)^2 + \cdot \cdot \cdot
\end{equation*}

and

\begin{equation*}
f^\prime(a) = c_1
\end{equation*}

- Continuing with the second derivative...

\begin{equation*}
f^{\prime \prime}(x) = 2 c_2 + 3 (2) c_3(x-a) + 4 (3) c_4(x-a)^3 +
\cdot \cdot \cdot
\end{equation*}

and 

\begin{equation*}
f^{\prime \prime}(a)= 2 c_2 \implies c_2 = \frac{f^{\prime \prime}(a)}{2}
\end{equation*}

- The third derivative gives

\begin{equation*}
f^{\prime \prime \prime}(x) = 3 (2) c_3 + 4 (3) (2) c_4(x-a) +
\cdot \cdot \cdot
\end{equation*}

and 

\begin{equation*}
f^{\prime \prime \prime}(a)= 3 (2) c_2 \implies c_3 = \frac{f^{\prime
    \prime \prime}(a)}{3 (2)}
\end{equation*}    
    
- Generalizing

\begin{equation*}
c_n = \frac{f^{(n)}(a)}{n!}
\end{equation*}

- Note that this works for $n=0$ since $0! =1$ and define $f^{(0)}(x) = f(x)$.

- The Taylor series is

\begin{eqnarray*}
f(x) &=& \sum \limits_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!} (x-a)^n\\
&=& f(a) + f^\prime(a) (x-a) + \frac{f^{\prime \prime} (a)}{2!}(x-a)^2+
    \frac{f^{\prime \prime \prime} (a)}{3!}(x-a)^3 + \cdot \cdot \cdot
\end{eqnarray*}

- Define the nth degree Taylor polynomial of $f(x)$ as

\begin{eqnarray*}
T_n(x) &=& \sum \limits_{i=0}^{n} \frac{f^{(i)}(a)}{i!} (x-a)^i\\
\end{eqnarray*}

- The remainder is defined as 

\begin{equation*}
R_n(x) = f(x) - T_n(x)
\end{equation*}

so the remainder is just the {\em error} between $f(x)$ and the nth
degree Taylor polynomial, i.e., 

\begin{equation*}
f(x) = T_n(x) + R_n(x)
\end{equation*}


```python
# Homework 1 due Jan. 20, 2020
#-----------------------------------------------------------------------------------------------
import numpy as np  #helps to abbreviate, otherwise the "handle" is numpy.
from matplotlib import pyplot as plt
import math as m
#-----------------------------------------------------------------------------------------------
def Tn_sin(n,x): #returns the nth degree Taylor polynomial for the sin function where n=degree.
    i = 0
    f = ((-1)**i)*(x**(2*i+1))/m.factorial(2*i+1)
    for i in range(1, n):
        f = f + ((-1)**i)*(x**(2*i+1))/m.factorial(2*i+1)
    return f
#-----------------------------------------------------------------------------------------------
def Tn_cos(n,x): #returns the nth degree Taylor polynomial for the cos function where n=degree.
    i = 0
    f = ((-1)**i)*(x**(2*i))/m.factorial(2*i)
    for i in range(1, n):
        f = f + ((-1)**i)*(x**(2*i))/m.factorial(2*i)
    return f
#-----------------------------------------------------------------------------------------------
def Tn_exp(n,x):  #returns the nth degree Taylor polynomial for the exp function where n=degree.
    i = 0
    f = x**i/m.factorial(i)
    for i in range(1, n):
        f = f + x**(i)/m.factorial(i)
    return f
#-----------------------------------------------------------------------------------------------
#beginning of main program

n = [2,3,5,7] #degree of Taylor polynomial
x = np.linspace(0,2,100)

fx = [Tn_exp(n[0],x),Tn_exp(n[1],x),Tn_exp(n[2],x),Tn_exp(n[3],x)]
fy = [Tn_sin(n[0],x),Tn_sin(n[1],x),Tn_sin(n[2],x),Tn_sin(n[3],x)]
fz = [Tn_cos(n[0],x),Tn_cos(n[1],x),Tn_cos(n[2],x),Tn_cos(n[3],x)]

#plot results.  Think of plt as an object that encapsulates a group of related procedures or "methods".
#Plot of 2nd degree Taylor polynimial of exp(x) compared to exp(x)
plt.plot(x,fx[0],'b')         #'b' = blue, you can also add symbols or change the linestyle.
plt.plot(x,np.exp(x),'r')     #'r' = red
plt.xlabel('x')
plt.ylabel('y')
plt.title('2nd degree Taylor polynomial')
plt.show()
#Plot of 3rd degree Taylor polynimial of exp(x) compared to exp(x)
plt.plot(x,fx[1],'b')         #'b' = blue, you can also add symbols or change the linestyle.
plt.plot(x,np.exp(x),'r')     #'r' = red
plt.xlabel('x')
plt.ylabel('y')
plt.title('3rd degree Taylor polynomial')
plt.show()
#Plot of 5th degree Taylor polynimial of exp(x) compared to exp(x)
plt.plot(x,fx[2],'b')         #'b' = blue, you can also add symbols or change the linestyle.
plt.plot(x,np.exp(x),'r')     #'r' = red
plt.xlabel('x')
plt.ylabel('y')
plt.title('5th degree Taylor polynomial')
plt.show()
#Plot of 7th degree Taylor polynimial of exp(x) compared to exp(x)
plt.plot(x,fx[3],'b')         #'b' = blue, you can also add symbols or change the linestyle.
plt.plot(x,np.exp(x),'r')     #'r' = red
plt.xlabel('x')
plt.ylabel('y')
plt.title('7th degree Taylor polynomial')
plt.show()
#----------------------------------------------------------------------------------------------------------------
#Plot of 2nd degree Taylor polynimial of sin(x) compared to sin(x)
plt.plot(x,fy[0],'b')         #'b' = blue, you can also add symbols or change the linestyle.
plt.plot(x,np.sin(x),'r')     #'r' = red
plt.xlabel('x')
plt.ylabel('y')
plt.title('2nd degree Taylor polynomial')
plt.show()
#Plot of 3rd degree Taylor polynimial of sin(x) compared to sin(x)
plt.plot(x,fy[1],'b')         #'b' = blue, you can also add symbols or change the linestyle.
plt.plot(x,np.sin(x),'r')     #'r' = red
plt.xlabel('x')
plt.ylabel('y')
plt.title('3rd degree Taylor polynomial')
plt.show()
#Plot of 5th degree Taylor polynimial of sin(x) compared to sin(x)
plt.plot(x,fy[2],'b')         #'b' = blue, you can also add symbols or change the linestyle.
plt.plot(x,np.sin(x),'r')     #'r' = red
plt.xlabel('x')
plt.ylabel('y')
plt.title('5th degree Taylor polynomial')
plt.show()
#Plot of 7th degree Taylor polynimial of sin(x) compared to sin(x)
plt.plot(x,fy[3],'b')         #'b' = blue, you can also add symbols or change the linestyle.
plt.plot(x,np.sin(x),'r')     #'r' = red
plt.xlabel('x')
plt.ylabel('y')
plt.title('7th degree Taylor polynomial')
plt.show()
#--------------------------------------------------------------------------------------------------------------
#Plot of 2nd degree Taylor polynimial of cos(x) compared to cos(x)
plt.plot(x,fz[0],'b')         #'b' = blue, you can also add symbols or change the linestyle.
plt.plot(x,np.cos(x),'r')     #'r' = red
plt.xlabel('x')
plt.ylabel('y')
plt.title('2nd degree Taylor polynomial')
plt.show()
#Plot of 3rd degree Taylor polynimial of cos(x) compared to cos(x)
plt.plot(x,fz[1],'b')         #'b' = blue, you can also add symbols or change the linestyle.
plt.plot(x,np.cos(x),'r')     #'r' = red
plt.xlabel('x')
plt.ylabel('y')
plt.title('3rd degree Taylor polynomial')
plt.show()
#Plot of 5th degree Taylor polynimial of cos(x) compared to cos(x)
plt.plot(x,fz[2],'b')         #'b' = blue, you can also add symbols or change the linestyle.
plt.plot(x,np.cos(x),'r')     #'r' = red
plt.xlabel('x')
plt.ylabel('y')
plt.title('5th degree Taylor polynomial')
plt.show()
#Plot of 7th degree Taylor polynimial of cos(x) compared to cos(x)
plt.plot(x,fz[3],'b')         #'b' = blue, you can also add symbols or change the linestyle.
plt.plot(x,np.cos(x),'r')     #'r' = red
plt.xlabel('x')
plt.ylabel('y')
plt.title('7th degree Taylor polynomial')
plt.show()
```

Given the Taylor series expansion,

\begin{eqnarray*}
f(x) &=& \sum \limits_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!} (x-a)^n\\
&=& f(a) + f^\prime(a) (x-a) + \frac{f^{\prime \prime} (a)}{2!}(x-a)^2+
    \frac{f^{\prime \prime \prime} (a)}{3!}(x-a)^3 + \cdot \cdot \cdot
\end{eqnarray*}

Problem #1:
Find the Taylor series for $f(x) = e^x$ about x = 0. Show that $$e^x = \sum \limits_{n=0}^{\infty} \frac{x^n}{n!}$$

First, let's find the first four terms. 

Let, $f^{(n)}(x) = e^x$ $\forall n \in \mathbb{N} $ and $a = 0$

First term:
$f(a) = 0 = \frac{x^0}{0!}$

Second term:
$f^\prime(a) (x-a) = (1)x = \frac{x^1}{1!}$

Third term:
$\frac{f^{\prime \prime} (a)}{2!}(x-a)^2 = \frac{1}{2!}(x^2) = \frac{x^2}{2!}$

Given that $e^x$ is completely smooth and if we follow the terms inductively, this shows (doesn't properly prove however) that $$e^x = \sum \limits_{n=0}^{\infty} \frac{x^n}{n!}$$

Problem #2: Find the Taylor series for $f(x) = sin(x)$ about x = 0. Show that $$sin(x) = \sum \limits_{n=0}^{\infty} \frac{(-1)^n x^{2n+1}}{(2n+1)!}$$

First, let's find the first four terms. 

Let, $f^{(n)}(x) = sin(x)$ $\forall n \in \mathbb{N} $ and $a = 0$

First term:
$f(a) = 0$

Second term:
$f^\prime(a) (x-a) = 1(x) = x = \frac{(-1)^0 x^{2(0) + 1}}{(2(0) + 1)!}$

Third term:
$\frac{f^{\prime \prime} (a)}{2!}(x-a)^2 = \frac{0}{2!}(x^2) = 0$

Fourth term:
$\frac{f^{\prime \prime \prime} (a)}{3!}(x-a)^3 = \frac{-1}{3!}(x^3) = \frac{(-1)^1 x^{2(1) + 1}}{(2(1) + 1)!}$

Given that $sin(x)$ is completely smooth and if we follow the terms inductively, this shows (doesn't properly prove however) that $$sin(x) = \sum \limits_{n=0}^{\infty} \frac{(-1)^n x^{2n+1}}{(2n+1)!}$$

Problem #3: Find the Taylor series for $f(x) = cos (x)$ about x = 0. Show that $$cos(x) = \sum \limits_{n=0}^{\infty} \frac{(-1)^n x^{2n}}{(2n)!}$$

First, let's find the first four terms. 

Let, $f^{(n)}(x) = sin(x)$ $\forall n \in \mathbb{N} $ and $a = 0$

First term:
$f(a) = 1 = \frac{(-1)^0 x^{2(0)}}{(2(0))!}$

Second term:
$f^\prime(a) (x-a) = 0(x) = 0$

Third term:
$\frac{f^{\prime \prime} (a)}{2!}(x-a)^2 = \frac{-1}{2!}(x^2) = \frac{(-1)^1 x^{2(1)}}{(2(1))!}$

Fourth term:
$\frac{f^{\prime \prime \prime} (a)}{3!}(x-a)^3 = \frac{0}{3!}(x^3) = 0$

Given that $cos(x)$ is completely smooth and if we follow the terms inductively, this shows (doesn't properly prove however) that $$cos(x) = \sum \limits_{n=0}^{\infty} \frac{(-1)^n x^{2n}}{(2n)!}$$
