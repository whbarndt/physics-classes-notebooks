# Lab 7: Planetary motion

- The motion of the Sun and Earth is a *two-body problem*.  
- Problem reduced to a one-body problem by: 1) noting the mass of
  the Sun is much greater than the mass of Earth, 2) assume the Sun is
  stationary and fix origin at Sun's center.
- One-body problem is more general.  Introduce reduced mass,
\begin{equation*}
\mu = \frac{mM}{m+M}
\end{equation*}
where m = $6 \times 10^{24}$ kg (Earth) and M = $2 \times 10^{30}$ kg
(Sun). Thus $\mu \approx m$.
- In the following, consider a single particle of mass $m$ moving
  about a fixed center of force, which we take as the origin of the
  coordinate system.
- Newton's universal law of gravitation is
\begin{equation*}
{\rm \bf F} = -\frac{GmM}{r^3} {\rm \bf r} 
\end{equation*}
where ${\rm \bf r}$ is the vector from $M$ to $m$ and $G=6.67 \times
10^{-11}$ m$^3$/(kg s$^2$).
- The force law applies to objects of negligible spatial extent,
  but also applies to any uniform sphere or spherical shell if the
  distance $r$ is measured from the center of each mass.
- Gravitational force has two general properties: 1) its magnitude
  depends only on the separation of the objects and 2) its direction is
  along the line joining the objects.  Such a force is called a
  *central force*. 
- The assumption of central force implies that the orbit (Earth)
  is restricted to a plane (e.g., $xy$) and the *angular momentum*
  is conserved and lies in the third direction ($z$), i.e., 
\begin{equation*}
L_z = ({\rm \bf r} \times m {\rm \bf v})_z = m (x v_y -y v_x)
\end{equation*}
Note that in general $ {\rm \bf L} = {\rm \bf r} \times {\rm \bf p}$ with
${\rm \bf p} = m {\rm \bf v}$.
- Energy conservation applies too, i.e., 
\begin{equation*}
E = \frac{1}{2} m v^2 - \frac{GmM}{r}
\end{equation*}

## Equations of motion

- The equation of motion is
\begin{equation*}
m \frac{d^2 {\rm \bf r}}{dt^2} = - \frac{GmM}{r^3}{\rm \bf r}
\end{equation*}
-  In force in Cartesian coordinates is
\begin{eqnarray*}
F_x &=& - \frac{GmM}{r^2} \cos \theta = -\frac{GmM}{r^3} x\\
F_y & = & -\frac{GmM}{r^2} \sin \theta = -\frac{GmM}{r^3} y\\
\end{eqnarray*}
- Thus the Cartisian equations of motion are
\begin{eqnarray*}
\frac{d^2 x}{dt^2} &=& - \frac{GM}{r^3}x\\
\frac{d^2 y}{dt^2} &=& - \frac{GM}{r^3}y
\end{eqnarray*}

## Circular orbits

- Planetary motion is nearly circular, so
\begin{equation*}
a = \frac{v^2}{r}
\end{equation*}
where $v$ is the speed of the object and $a$ is directed towards the
center and is due to the gravitational force, i.e.,
\begin{equation*}
\frac{mv^2}{r} = \frac{GmM}{r^2}
\end{equation*}
or \begin{equation*}
v = \left( \frac{GM}{r} \right)^{1/2}
\end{equation*}
- The period of the orbit is
\begin{equation*}
T = \frac{2 \pi r}{v}
\end{equation*}
and 
\begin{equation*}
T^2 = \frac{4 \pi^2 r^3}{GM}
\end{equation*}
-  This is a special case of Kepler's third law where $r$
  corresponds to the semi-major axis of an ellipse.


## Elliptical orbits

- The most general orbit is an ellipse.
- The two foci of an ellipse, $F_1$ and $F_2$, have the property
  that for any point $P$ the distance $F_1P + F_2 P = const$.
- The longer axis is the major axis and half this length is the
  *semi-major axis* $a$.
- The shorter axis is the minor axis and half this length is the
  *semi-minor axis* $b$.
 Specify elliptical orbit by $a$ and the *eccentricity* $e$
  where $e$ is the ratio of the distance between the foci to the
  length of the major axis.
- Since $F_1P + F_2P = 2a$, 
\begin{equation*}
e = \sqrt{1 - \frac{b^2}{a^2}}
\end{equation*}
with $0 < e < 1$.  For a circle $e=0$.  For Earth's orbit, $e = 0.0167$.

## Astronomical Units

- Since large numbers (once upon a time) are awkward, it is
  convenient to choose a system of units in which the magnitude of the
  produce $GM$ is the order of unity.
- Convenient unit of length is the *astronomical unit* with
\begin{equation*}
1 {~\rm AU} = 1.496 \times 10^{11} {~{\rm m}}
\end{equation*} 
-  The unit of time is one year = $3.15 \times 10^7$ s.  
-  In these units, $T$ = one year, $a$ = 1 AU, 
\begin{equation*}
GM = \frac{4 \pi^2 a^3}{T^2} = 4 \pi^2 {\rm AU^3/yr^2}
\end{equation*}

## Mini solar system

- The presence of other planets implies that the total force on a
  planet is no longer a central force.
- Furthermore, the orbits of the planets are not exactly in the
  same plane, so the analysis must be extended to three dimensions.
- For simplicity, consider two planets of mass $m_1$ and $m_2$.
  The equations of motion are
\begin{eqnarray*}
m_1 \frac{d^2 {\rm \bf r}_1}{dt^2} &=& - \frac{m_1G  M}{r_1^3} {\rm \bf r}_1 +
  \frac{m_1Gm_2}{r_{21}^3} {\rm \bf r}_{21} \\
m_2 \frac{d^2 {\rm \bf r}_2}{dt^2} &=& - \frac{m_2G  M}{r_2^3} {\rm \bf r}_2 -
  \frac{m_2Gm_1}{r_{21}^3} {\rm \bf r}_{21} \\
\end{eqnarray*}
where ${\rm \bf r}_1$ and ${\rm \bf r}_2$ are directed from the Sun to the
planets and ${\rm \bf r}_{21} = {\rm \bf r}_2 - {\rm \bf r}_1$.
- No analytical solution exists!  Requires a numerical solution.


## Introduction to Python "classes".

Classes provide a means of bundling data and functionality together. Creating a new class creates a new type of object, allowing new instances of that type to be made. Each class instance can have attributes attached to it for maintaining its state. Class instances can also have methods (defined by its class) for modifying its state.

Planetary motion is the perfect application for Python classes.  Here we want to bundle the properties of a given planet into a single object that contains data and methods.

````python
import numpy as np
import matplotlib.pyplot as plt

GM = 4*np.pi**2

class planet:
    
    """
    define a class planet that contains initial position and velocity as well as mass and color.
    The class is called with arguments (x0,y0,vx0,vy0,m,color)
    """
    
    def __init__(self,x0,y0,vx0,vy0,m,clr):
        self.xn = x0
        self.yn = y0
        self.vxn = vx0
        self.vyn = vy0
        self.m = m
        self.clr = clr
        self.E0 = 0.5*np.sqrt(vx0**2+vy0**2) - (GM)/np.sqrt(x0**2 + y0**2)
        self.L0 = self.m*(x0*vy0 - y0*vx0)
        self.x = []
        self.y = []
        self.vx = []
        self.vy = []
        self.time = []
        self.energy = []
        self.angular_momentum = []
        self.areas = []

    def F(self,r,x,v):
        return np.array([v,-GM*x/r**3])
    
    def rk4_step(self,r,x,v):
        k1 = dt*self.F(r,x,v)         
        k2 = dt*self.F(r,x + 0.5*k1[0], v + 0.5*k1[1])
        k3 = dt*self.F(r,x + 0.5*k2[0], v + 0.5*k2[1])
        k4 = dt*self.F(r,x + k3[0], v + k3[1])
        x = x + (k1[0] + 2*k2[0] + 2*k3[0] + k4[0])/6.
        v = v + (k1[1] + 2*k2[1] + 2*k3[1] + k4[1])/6.
        return x,v
        
    def move_planet(self,tfinal,dt): 
        t = 0.0
        while t < tfinal+dt:
            self.time.append(t)
            self.x.append(self.xn)
            self.y.append(self.yn)
            self.vx.append(self.vxn)
            self.vy.append(self.vyn)
            self.energy.append(self.calculate_energy())
            self.angular_momentum.append(self.calculate_angular_momentum())
            r = np.sqrt(self.xn**2 + self.yn**2)
            self.xn,self.vxn = self.rk4_step(r,self.xn,self.vxn)  #x-component
            self.yn,self.vyn = self.rk4_step(r,self.yn,self.vyn)  #y-component
            t = t + dt

    def plot_planet(self,nstep):
        fig = plt.figure()
        ax = fig.add_subplot(111)
        ax.plot(self.x[::nstep],self.y[::nstep],'.',markersize=0.5,color=self.clr)
        ax.set_xlabel('x (AU)')
        ax.set_ylabel('y (AU)')
        ax.set_aspect('equal')
        ax.plot([0,0],[0,0],'o',markersize=10.0,color='gold')
        ax.minorticks_on()
        ax.grid(True, which='both')
        plt.show()
            
    def calculate_energy(self):
        return (0.5*np.sqrt(self.vxn**2+self.vyn**2)**2) - (GM)/np.sqrt(self.xn**2 + self.yn**2)

    def plot_energy_conservation(self):
        fig = plt.figure()
        ax = fig.add_subplot(111)
        delta_energy = np.abs(np.divide(np.subtract(self.energy,self.E0),self.E0))
        ax.plot(self.time,delta_energy,'.',markersize=0.5)
        ax.set_ylabel('Energy (J)')
        ax.set_xlabel('Time (yrs)')
    
    def calculate_angular_momentum(self):
        return self.m*(self.xn*self.vyn - self.yn*self.vxn)
    
    def plot_ang_momentum_conservation(self):
        fig = plt.figure()
        ax = fig.add_subplot(111)
        delta_momentum = np.abs(np.divide(np.subtract(self.angular_momentum,self.L0),self.L0))
        ax.plot(self.time, delta_momentum,'.',markersize=0.5)
        ax.set_ylabel('Angular Momentum (AU*m^2/yr)')
        ax.set_xlabel('Time (yrs)')
    
    def verify_keplers_second_law(self):
      for i in range(5):
        init_theta = np.arctan2(self.y[i*100], self.x[i*100])
        final_theta = np.arctan2(self.y[100+i*100], self.x[100+i*100])
        delta_theta = final_theta - init_theta
        sum_r = 0;
        for j in range(100):
          sum_r += np.sum(np.sqrt(np.square(self.x[j+i*100])+np.square(self.y[j+i*100])))
        avg_r = sum_r/100
        self.areas.append(1/2*(avg_r**2)*delta_theta)
````

````python
#Class Solar System.

class solar_system:

    '''
    defines a solar system class that you pass planet objects in
    '''

    def __init__(self, *args):
        self.num_planets = len(args)
        self.planets = []
        for pl in args:
          self.planets.append(pl)

    def F_m(self,eff_mass,rn,sn,vn,delta_r,delta_s):
        F = 0
        #Adding up all the forces effecting the current planet
        for i in range(len(eff_mass)):
          F += GM*eff_mass[i]*delta_s[i]/(delta_r[i]**3)
        return np.array([vn,-(GM*sn)/(rn**3) - F]) #Subtracting the added forces from Sun force
    
    def rk4_step(self,eff_mass,rn,sn,vn,delta_r,delta_s):
        k1 = dt*self.F_m(eff_mass,rn,sn,vn,delta_r,delta_s)         
        k2 = dt*self.F_m(eff_mass,rn,sn + 0.5*k1[0],vn + 0.5*k1[1], delta_r, delta_s)
        k3 = dt*self.F_m(eff_mass,rn,sn + 0.5*k2[0],vn + 0.5*k2[1], delta_r, delta_s)
        k4 = dt*self.F_m(eff_mass,rn,sn + k3[0],vn + k3[1], delta_r, delta_s)
        sn += (k1[0] + 2*k2[0] + 2*k3[0] + k4[0])/6.
        vn += (k1[1] + 2*k2[1] + 2*k3[1] + k4[1])/6.
        return sn,vn
        
    def move_planets(self,tfinal,dt): 
        t = 0
        while t < tfinal+dt:
            for current_planet in self.planets:
              current_planet.time.append(t)
              current_planet.x.append(current_planet.xn)
              current_planet.y.append(current_planet.yn)
              current_planet.vx.append(current_planet.vxn)
              current_planet.vy.append(current_planet.vyn)
              current_planet.energy.append(current_planet.calculate_energy())
              current_planet.angular_momentum.append(current_planet.calculate_angular_momentum())
              rn = np.sqrt(current_planet.xn**2 + current_planet.yn**2)
              x_diffs = []
              y_diffs = []
              r_diffs = []
              eff_mass = []
              for others in self.planets:
                if current_planet != others:
                  delta_x = current_planet.xn - others.xn
                  delta_y = current_planet.yn - others.yn
                  x_diffs.append(delta_x)
                  y_diffs.append(delta_y)
                  r_diffs.append(np.sqrt(delta_x**2 + delta_y**2))
                  eff_mass.append(others.m)
              current_planet.xn,current_planet.vxn = self.rk4_step(eff_mass,rn,current_planet.xn,current_planet.vxn,x_diffs,r_diffs)  #x-component
              current_planet.yn,current_planet.vyn = self.rk4_step(eff_mass,rn,current_planet.yn,current_planet.vyn,y_diffs,r_diffs)  #y-component
            t += dt

    def plot_planets(self,nstep):
      for pl in self.planets:
        plt.plot(pl.x,pl.y,'.',markersize=0.5,color=pl.clr)
      plt.xlabel('x (AU)')
      plt.ylabel('y (AU)')
      plt.axes().set_aspect('equal')
      plt.plot([0,0],[0,0],'o',markersize=10.0,color='gold')
      plt.show()
      for pl in self.planets:
        pl.plot_energy_conservation()
        pl.plot_ang_momentum_conservation()
````
