```python
import numpy as np
import matplotlib.pyplot as plt

# Variables
dx = 0.001
x = np.arange(0,1,dx)

# R-Values
dr = 0.001
r_values = np.arange(0,1,dr)

# Calculated Lyaponov
lyap_values = []

# Lyaponov Calculation
for i in range(len(r_values)):
  lyap_exp = 0
  for j in range(len(x)):
    lyap_exp += np.log(r_values[i]*np.pi*np.cos(np.pi*x[j]))
  lyap_exp = lyap_exp/len(x)
  lyap_values.append(lyap_exp)

# Plotting
plt.figure()
plt.plot(r_values,lyap_values)
plt.title("Lyapunov Exponent")
plt.xlabel('Parameter (r)')
plt.ylabel('$\\lambda$')
```
